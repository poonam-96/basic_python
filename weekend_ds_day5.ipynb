```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# If we are using Matplotlib from within a script then the plt.show() command is of great use
# It starts an event loop, looks for all currently active figure objects and opens one or more interactive windows that 
# display the figure or figures
```

```python
x1 = np.linspace(0,10,100)
```

```python
x1
```

```python
plt.plot(x1,np.sin(x1))
plt.plot(x1,np.cos(x1),'--')
```

```python
np.sin(x1)
```

```python
plt.subplot(2,1,1)
plt.plot(x1,np.sin(x1))

plt.subplot(2,1,2)
plt.plot(x1,np.cos(x1))
plt.show()
```

```python
plt.plot([1,2,3,4])
plt.ylabel('Numbers')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('Numbers')
plt.ylabel('Squared Numbers')
plt.show()
```

```python
x = np.linspace(0,2,100)

plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.legend()
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Simple Plot')

plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,16])
plt.show()
```

```python
t = np.arange(0.,5.,0.2)
```

```python
t
```

```python
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
t = np.linspace(0.,5.,100)
```

```python
# First create a grid of plots
# ax will be an array of two axes objects
fig,ax = plt.subplots(2)
ax[0].plot(t,np.sin(t),'b-')
ax[1].plot(t,np.cos(t),'b-')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
x3 = np.arange(0,6,0.01)
```

```python
plt.plot(x3,[xi**2 for xi in x3],'b-')
plt.show()
```

# Scatter Plot

```python
x7 = np.linspace(0,10,30)
y7 = np.sin(x7)
plt.plot(x7,y7,'o',color='black')
plt.show()
```

```python
data1 = np.random.randn(1000)
```

```python
plt.hist(data1)
plt.show()
```

```python
data2 = [5,25,50,20]
plt.bar(range(len(data2)),data2)
plt.show()
```

```python
range(len(data2))
```

```python
data2 = [5,25,50,20]
plt.barh(range(len(data2)),data2)
plt.show()
```

```python
A = [15,30,45,22]
B = [15,25,50,20]
z2 = range(4)
plt.bar(z2,A,color='b')
plt.bar(z2,B,color='r',bottom=A)
plt.show()
```

```python
plt.figure(figsize=(7,7))
x10 = [35,25,20,20]
labels = ['computer','electronics','mechanical','chemical']
plt.pie(x10,labels=labels)
plt.show()
```
